```python
import wikipedia
ny = wikipedia.page(" Amitabh  Bachan")
ny.content
```

```python
import nltk
nltk.download('punkt_tab')
```

```python
from MultiExtractiveSummarizer import MultiExtractiveSummarizer

# Initialize the summarizer
summarizer = MultiExtractiveSummarizer(embedding_method='tfidf', summarization_method='lexrank')

# Example text document
text = ny.content

# Generate the summary with number of sentences
summary = summarizer.summarize(text, num_sentences=20)

print("Summary:")
print(summary)

# Generate the summary ratio of text
#summary = summarizer.summarize(text, ratio=0.1)

#print("Summary:")
#print(summary)
```
